```python
# Splitting who links in separate samples
dataset["Ссылка на сайт"].loc[0] = "http://greenstartup.ru/"
dataset.loc[93] = dataset.loc[0]
dataset["Ссылка на сайт"].loc[93] = "http://9911850.ru/"

# Adding http prefix 
def fix_url(url):
    if not isinstance(url, str):
        return None

    if not url.startswith("http"):
        return "http://"+url
    
    return url
    
dataset["Ссылка на сайт"] = dataset["Ссылка на сайт"].apply(fix_url)
dataset["Ссылка на соц сеть"] = dataset["Ссылка на соц сеть"].apply(fix_url)
```

```python
dataset.dropna(axis=0, subset=["Ссылка на сайт"], inplace=True)
```
